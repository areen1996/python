```python
import random
playerConstant = 'O'
computerConstant = 'X'
board=[1,2,3,4,5,6,7,8,9]
  
def DisplayBoard(board):
    print("+-------+-------+-------+")
    print("|       |       |       |")
    print("|   {b[0]}   |   {b[1]}   |   {b[2]}   |".format(b=board))
    print("|       |       |       |")
    print("+-------+-------+-------+")
    print("|       |       |       |")
    print("|   {b[3]}   |   {b[4]}   |   {b[5]}   |".format(b=board))
    print("|       |       |       |")
    print("+-------+-------+-------+")
    print("|       |       |       |")
    print("|   {b[6]}   |   {b[7]}   |   {b[8]}   |".format(b=board))
    print("|       |       |       |")
    print("+-------+-------+-------+")

    

def EnterUserMove(board):
    '''Asks the user about their move, checks the input, and updates the board according to the user's decision'''
    user=int(input("enter O in the position 1-9 :"))
    if board[user-1]!="X" and board[user-1]!="O":   
        board[user-1]="O"
    else:
        while True:
            print("choose another position :")
            user=int(input("enter O in the position 1-9 :"))
            if board[user-1]!="X" and board[user-1]!="O":   
                board[user-1]="O"
                break
        
def DrawComputerMove(board):
    '''The function draws the computer's move and updates the board'''
    while True:    
        computer = random.randint(1,9)
        if board[computer-1]!="X" and board[computer-1]!="O":
            board[computer-1]="X"
            break
    
    
def MakeListOfFreeFields(board):
    '''The function browses the board and builds a list of all the free squares'''
    b=[]
    for i in range(len(board)):
        if board[i]!="X" and board[i]!="O":
            b.append(i)
    print(b)
    if b==[]:
        return False
    else:
        return True
        
def VictoryFor(board,sign):
    '''the function analyzes the board status in order to check if the player using 'O's or 'X's has won the game''' 
    for i in range(0,len(board),3):
        if board[i] == board[i+1] == board[i+2]==sign :
            return True
    for i in range(3):
        if board[i]== board[i+3] == board[i+6]==sign:
            return True
    if board[0]==board[4]==board[8]=="X":
        return True
    if board[2]==board[4]==board[6]=="X":
        return True
    else :
        return False
        
    
while True:
    DisplayBoard(board)
    if (MakeListOfFreeFields(board)) :
        DrawComputerMove(board)
        DisplayBoard(board)
        if VictoryFor(board,"X") :
            print("X win ")
            break
        elif VictoryFor(board,"O") :
            print("O win ")
            break
        elif (MakeListOfFreeFields(board) ):
            EnterUserMove(board)
            if VictoryFor(board,"O") :
                print("O win ")
                break
            elif VictoryFor(board,"O") :
                print("O win ")
                break
            else :
                MakeListOfFreeFields(board)
                

    
```
